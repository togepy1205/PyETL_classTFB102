```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/movie/index.html'
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
headers = {
    'User-Agent': userAgent
}

res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser') # 'html.parser' 這欄為指定轉換的格式, 另有其他不同的
                                              # lxml -> pip install lxml (另一種轉換器)
titleTagList = soup.select('div.title')
for titleTag in titleTagList:
    try:
        title = titleTag.select('a')[0].text # 若有回傳空值, 會導致 Error -> use try except
        # titleTag(Tag Obj.), .select('a')(list),[0](取值),.txet(取文字)
        articleUrl = 'https://www.ptt.cc' + titleTag.select('a')[0]['href']
        print(title)
        print(articleUrl)
    except IndexError as e:
        #print(e) # 錯誤資訊不一定要 print 出來
        print(titleTag)
    print('==================')
```
