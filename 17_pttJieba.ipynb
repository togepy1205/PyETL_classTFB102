```python
import jieba
import os

jieba.load_userdict('./mydict.txt')
stopWords = list()
with open ('./stopword.txt', 'r', encoding='utf-8') as f:
    stopWords = [w.replace('\n', '') for w in f.readlines()] # ['a\n, 'b\n,.....]
#os.listdir('./res_gossiping') # 將目錄裡的檔案名稱用 list 呈現
sb = ''
for fileName in os.listdir('./res_gossiping'):
    filePath = './res_gossiping/' + fileName
    #print(filepath)
    try:
        with open(filePath, 'r', encoding = 'utf-8') as f:
            sb += f.read().split('---split---')[0]
    except IndexError:
        pass
    except PermissionError:
        pass
    
wordCut = jieba.cut(sb)
wordCutList = [w for w in wordCut] # 因為原本的出來為 generator, 但想要留下資料, 便用 list 把資料留起來
#for w in wordCut:
#    print(w)

wordCutList
wordCountDict = dict()
for w in wordCutList:
    if (w.__len__() < 2) or (w in stopWords):
        continue
    if w in wordCountDict:
        wordCountDict[w] += 1
    else:
        wordCountDict[w] = 1
        
wordCountDict
wordCountList = [(k, v) for k,v in wordCountDict.items()]
wordCountList.sort(key=lambda x: x[1], reverse=True) # 用 tuple 的第一個值做排序, 做降冪
wordCountList
```
