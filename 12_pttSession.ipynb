```python
# 如何被綁 cookies 的過程, 並且 session 如何存他們下來
import requests
from bs4 import BeautifulSoup

urlFirst = 'https://www.ptt.cc/ask/over18?from=%2Fbbs%2FGossiping%2Findex.html' # ptt 18 禁的警告頁面
urlSecond = 'https://www.ptt.cc/ask/over18' # 按下同意時的跳轉頁面 (看不到 , 但是該頁面跳轉瞬間被綁 cokkies)
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
headers = {
    'User-Agent': userAgent
}

ss = requests.session() # 建立 session 來儲存 cookies
print(ss.cookies)
print('============')

resFirst = ss.get(urlFirst, headers=headers)
print(ss.cookies)
print('============')
soupFirst = BeautifulSoup(resFirst.text, 'html.parser')

data = dict() # 處理 R18 警告頁面的 post 用
data[soupFirst.select('input')[0]['name']] = soupFirst.select('input')[0]['value'] # 取出要去的網站的網址
data[soupFirst.select('button')[0]['name']] = soupFirst.select('button')[0]['value'] # 處理同意與不同意的按鈕, 所以取第 0 個
#data

# 按下同意按鈕的動作 (把資料帶進去)
ss.post(urlSecond, headers=headers, data=data) # 只是要路過頁面而已, 不需要宣告變數去接資料
print(ss.cookies)
print('============')

res = ss.get(url, headers=headers)
print(res.text)
```
