```python
import pandas as pd
import glob

data = list()
for filePath in glob.glob('./res_gossiping/*.txt'):
#     print(filePath)
    with open(filePath, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')[1:-1] # 用 split 切到剩你要的部分
        data.append(tmpList)
columns = [r.split(': ')[0] for r in data[0]]

df = pd.DataFrame(data=data, columns=columns)
df
```

```python
# 製作一個過濾器
def columnFilter(x):
    output = x.split(': ')[1]
    return output
```

```python
df['推'].apply(columnFilter) # 一維資料格式
```

```python
df
```

```python
# 備份 df, 因 pandas 很會重複使用記憶體, 如果不用 copy, df2 會因為 df 改變而跟著改變
df2 = df.copy()
```

```python
for c in df2:
    print(c)
```

```python
# 把原本冗字去掉
for c in df2:
    df2[c]= df[c].apply(columnFilter)
df2
```

```python
# 過濾邏輯不難時, 可用 lambda
df['推'].apply(lambda x: x.split(': ')[1])
```

```python
# 若出現 IndexError 代表有某一筆資料出問題了, 這時不要用 lambda, 請改用函數, 或是, 代表資料沒有洗乾淨 (建議從前面處裡)
for c in df:
    df[c] = df[c].apply(lambda x: x.split(': ')[1])
df
```

```python
# 把前面自動生成的 index 取消, header=None 把欄位取消, 預設為 utf-8 進 excel 會有問題, 須加上 sig
df.to_csv('./ptt.csv', index=False, header=None, encoding='utf-8-sig')
```

```python
# 讀 csv 時預設第一列會有欄位, 但有時資料本身完全沒有, 可將其取消並放進自己需要的
df3 = pd.read_csv('./ptt.csv', header=None)
df3.columns = columns
df3
```
