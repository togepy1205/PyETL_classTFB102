```python
# 將 PTT 文章列表上的文章爬下來並另存成文章
import requests
from bs4 import BeautifulSoup
import time
import random
import os #建立放文章的目錄用

if not os.path.exists('./pttMovie'):  # 檢查路徑是否存在 (避免重複創建造成錯誤)
    os.mkdir('./pttMovie') # 建立 line 43 放文章的資料夾

url = 'https://www.ptt.cc/bbs/movie/index.html'
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
headers = {
    'User-Agent': userAgent
}

for i in range(0, 5):
    res = None # 宣告在此
    while True:
        try:
            res = requests.get(url, headers=headers) # 變區域變數了, 所以在外面宣告一次
            break
        except:
            continue
            
    soup = BeautifulSoup(res.text, 'html.parser')

    titleTagList = soup.select('div.title')
    for titleTag in titleTagList:
        try:
            title = titleTag.select('a')[0].text 
            articleUrl = 'https://www.ptt.cc' + titleTag.select('a')[0]['href']
            # request article url
            articleUrl = 'https://www.ptt.cc/bbs/movie/M.1620790595.A.321.html'
            articleRes = requests.get(articleUrl, headers=headers)
            articleSoup = BeautifulSoup(articleRes.text, 'html.parser')
            #articleContent = articleSoup.select('div[id="main-content"]')[0].text.split('※ 發信站')[0]
            articleContentTag = articleSoup.select('div[id="main-content"]')[0] # 找到文章所在的位置
            for t in ['span', 'div']: # 把帶有 span, div 標籤的移除
                for tag in articleContentTag.select('t'): # 搜尋的標籤
                    tag.extract() # 把搜到的移除
            articleContent = articleContentTag.text # 接住移除後剩下的文字 (原本上面註解掉的部分)
            try: # 防止因標題出現日期常用的 / 導致讀取錯誤, 利用 try 去處理錯誤
                with open('./pttMovie/{}.txt'.format(title), 'w', encoding='utf-8') as f: # 勇 with 打開, 不用再寫關, 會自己幫你關, 然後將文章存下來放在文字檔裡
                    f.write(articleContent)
            except FileNotFoundError as e:
                with open('./pttMovie/{}.txt'.format(title.replace('/', '-')), 'w', encoding='utf-8') as f: # 將標題出錯的 / 變成 -
                    f.write(articleContent)
            except OSError as e: # 接住因出現非法字元造成的問題 (?等不能當檔案名的特殊字元)
                pass
            print(title)
            print(articleUrl)
        except IndexError as e:
            #print(e)
            print(titleTag)
        print('==================')

    newUrl = 'https://www.ptt.cc' + soup.select('a[class="btn wide"]')[1]['href'] 
    #print(newUrl)
    url = newUrl #因為要去上一頁, 所以要將原網址覆蓋掉
    
    #time.sleep(random.randint(1, 10)) # 隨機讓程式休息幾秒再繼續, 規避伺服器查驗爬蟲
```
